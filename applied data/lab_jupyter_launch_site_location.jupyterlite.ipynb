Next, let's create markers for all launch records.
If a launch was successful `(class=1)`, then we use a green marker and if a launch was failed, we use a red marker `(class=0)`


Note that a launch only happens in one of the four launch sites, which means many launch records will have the exact same coordinate. Marker clusters can be a good way to simplify a map containing many markers having the same coordinate.


Let's first create a `MarkerCluster` object


```python
marker_cluster = MarkerCluster()
```

*TODO:* Create a new column in `launch_sites` dataframe called `marker_color` to store the marker colors based on the `class` value


```python

# Create a map centered around a specific location
site_map = folium.Map(location=nasa_coordinate, zoom_start=4)

# Create a MarkerCluster object
marker_cluster = MarkerCluster()

# Define a function to determine marker color based on launch outcome
def determine_marker_color(outcome):
    if outcome == 1:
        return 'green'  # Success
    else:
        return 'red'  # Failure

# Iterate through the launch records
for index, row in spacex_df.iterrows():
    site = row['Launch Site']
    lat = row['Lat']
    long = row['Long']
    outcome = row['class']
    
    # Determine the color of the marker based on the launch outcome
    color = determine_marker_color(outcome)
    
    # Create a marker for the launch record
    folium.Marker(
        location=[lat, long],
        popup=f"Launch Site: {site}, Outcome: {'Success' if outcome == 1 else 'Failure'}",
        icon=folium.Icon(color=color)
    ).add_to(marker_cluster)

# Add the MarkerCluster layer to the map
site_map.add_child(marker_cluster)

# Display the map
site_map
```

*TODO:* For each launch result in `spacex_df` data frame, add a `folium.Marker` to `marker_cluster`


```python
# Iterate through the rows of spacex_df DataFrame
for index, row in spacex_df.iterrows():
    # Determine the launch outcome and set marker color accordingly
    if row['class'] == 1:
        marker_color = 'green'  # Success
    else:
        marker_color = 'red'  # Failure
    
    # Create a Marker object with its coordinate
    marker = folium.Marker(
        location=[row['Lat'], row['Long']],  # Coordinates
        popup=f"Launch Site: {row['Launch Site']}, Outcome: {'Success' if row['class'] == 1 else 'Failure'}",  # Popup message
        icon=folium.Icon(color=marker_color)  # Marker color
    )
    
    # Add the marker to the MarkerCluster
    marker_cluster.add_child(marker)

# Add the MarkerCluster to the site map
site_map.add_child(marker_cluster)

# Display the map
site_map
```

```python
from math import radians, sin, cos, sqrt, atan2

def calculate_distance(lat1, lon1, lat2, lon2):
    # Convert latitude and longitude from degrees to radians
    lat1_rad, lon1_rad = radians(lat1), radians(lon1)
    lat2_rad, lon2_rad = radians(lat2), radians(lon2)
    
    # Radius of the Earth in kilometers
    R = 6371.0
    
    # Calculate differences in latitude and longitude
    dlat = lat2_rad - lat1_rad
    dlon = lon2_rad - lon1_rad
    
    # Haversine formula
    a = sin(dlat / 2)**2 + cos(lat1_rad) * cos(lat2_rad) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))
    distance = R * c
    
    return distance

# Example usage:
distance = calculate_distance(29.559684888503615, -95.0830971930759, 28.5721, -80.648)
print("Distance between NASA Johnson Space Center and Cape Canaveral Space Launch Complex 40:", distance, "km")
```

```python
from math import sin, cos, sqrt, atan2, radians

def calculate_distance(lat1, lon1, lat2, lon2):
    # approximate radius of earth in km
    R = 6373.0

    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return distance
```
